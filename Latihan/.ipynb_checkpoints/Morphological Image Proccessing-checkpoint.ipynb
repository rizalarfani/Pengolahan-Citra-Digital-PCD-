# Operasi Dilasi A⊕B

```python
import cv2
import numpy as np

# Lokasi gambar anda
path = "../image/morfologi.png" 
# Membaca gambar
image = cv2.imread(path)

# Ubah menjadi layer 1 warna
# Dengan mengunakan cv2.cvtColor(image,COLOR_BGR2_GRAY)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
# Lihat ukuran image nya
print(gray.shape)


# Structering element
kernel = np.ones((3,3),np.uint8)
print(kernel)


# Lakulan opersai dilasi
dilasi = cv2.dilate(gray,kernel,iterations = 1)

# Menggunakan Metode cv2.imshow()
# untuk diplay gambar
cv2.imshow('Citra Asli',image)
cv2.imshow('Citra Dilasi',dilasi)
cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```

# Operasi Dilasi AΘB

```python
import cv2
import numpy as np

# Lokasi gambar anda
path = "../image/morfologi.png" 
# Membaca gambar
image = cv2.imread(path)

# Ubah menjadi layer 1 warna
# Dengan mengunakan cv2.cvtColor(image,COLOR_BGR2_GRAY)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
# Lihat ukuran image nya
print(gray.shape)


# Structering element
kernel = np.ones((5,5),np.uint8)
print(kernel)


# Lakulan opersai dilasi
dilasi = cv2.erode(gray,kernel,iterations = 1)

# Menggunakan Metode cv2.imshow()
# untuk diplay gambar
cv2.imshow('Citra Asli',image)
cv2.imshow('Citra Erosi',dilasi)
cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```

# Opersai Opening

```python
import cv2
import numpy as np

# Lokasi gambar anda
path = "../image/morfologi.png" 
# Membaca gambar
image = cv2.imread(path)

# Ubah menjadi layer 1 warna
# Dengan mengunakan cv2.cvtColor(image,COLOR_BGR2_GRAY)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
# Lihat ukuran image nya
print(gray.shape)


# Structering element
kernel = np.ones((5,5),np.uint8)
print(kernel)

# Lakulan opersai Opening
opening_image = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)

# Menggunakan Metode cv2.imshow()
# untuk diplay gambar
cv2.imshow('Citra Asli',image)
cv2.imshow('Citra Opening',opening_image)
cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```

# Operasi Closing

```python
import cv2
import numpy as np

# Lokasi gambar anda
path = "../image/morfologi.png" 
# Membaca gambar
image = cv2.imread(path)

# Ubah menjadi layer 1 warna
# Dengan mengunakan cv2.cvtColor(image,COLOR_BGR2_GRAY)
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
# Lihat ukuran image nya
print(gray.shape)


# Structering element
kernel = np.ones((5,5),np.uint8)
print(kernel)

# Lakulan opersai Opening
closing_image = cv2.morphologyEx(gray, cv2.MORPH_CLOSE, kernel)

# Menggunakan Metode cv2.imshow()
# untuk diplay gambar
cv2.imshow('Citra Asli',image)
cv2.imshow('Citra Closing',closing_image)
cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```
