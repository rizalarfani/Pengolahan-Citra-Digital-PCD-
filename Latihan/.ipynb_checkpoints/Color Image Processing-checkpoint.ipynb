# buat autoload library dan gambar dalam fungsi

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np


def loadgambar():
    return(cv2.imread('../image/lena.jpg'))

def tampil_plot(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


def figure():
    return plt.figure(num=None, figsize=(15, 20), dpi=100)
```

# CITRA RGB
### rumus color channel
#### b = img[:,:,0]

##### g = img[:,:,1]

##### r = img[:,:,2

```python
image = loadgambar()
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
r = image.copy()
# set green and blue channels to 0
r[:, :, 1] = 0
r[:, :, 0] = 0

g = image.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

b = image.copy()
# set red and green channels to 0
b[:, :, 1] = 0
b[:, :, 2] = 0

color = ('r', 'g', 'b')
plt.figure(num=None, figsize=(15, 20), dpi=100)
for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.subplot(4, 2, 1)
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    plt.title('Histogram citra asli')
plt.subplot(4, 2, 2), plt.imshow(image), plt.title('citra asli')

color = ('r')
for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.subplot(4, 2, 3)
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    plt.title('Histogram citra red')
plt.subplot(4, 2, 4), plt.imshow(tampil_plot(r)), plt.title('red')

color = ('g')
for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.subplot(4, 2, 5)
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    plt.title('Histogram citra green')
plt.subplot(4, 2, 6), plt.imshow(tampil_plot(g)), plt.title('green')

color = ('b')
for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.subplot(4, 2, 7)
    plt.plot(histr, color=col)
    plt.xlim([0, 256])
    plt.title('Histogram citra blue')
plt.subplot(4, 2, 8), plt.imshow(tampil_plot(b)), plt.title('blue')
plt.show()  # To show figure
```

# Intensity Adjustment

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
  
# membaca gambar di directory yang udah ditetukan
img = cv2.imread('../image/lena.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  
# Apply log transform.
c = 255/(np.log(1 + np.max(img)))
log_transformed = c * np.log(1 + img)
  
# Specify the data type.
log_transformed = np.array(log_transformed, dtype = np.uint8)

# untuk diplay gambar
cv2.imshow('Citran Asli',img)
cv2.imshow('Citra Adjustment',log_transformed)

cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```
