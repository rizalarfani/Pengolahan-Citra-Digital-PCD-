```python
import cv2
import numpy as np
import os

path = "../UTS/Bahan/segmentasi.jpg"
image = cv2.imread(path)
blur = cv2.medianBlur(image,(5))

lower_hijau = np.array([0,100,0])
uper_hijau = np.array([80,255,100])

mask = cv2.inRange(blur,lower_hijau,uper_hijau)
segmentasi = cv2.bitwise_and(blur,blur,mask = mask)

cv2.imshow("Citra Asli",image)
cv2.imshow("Citra Blur",blur)
cv2.imshow("Citra Segmentasi",segmentasi)
cv2.waitKey(0)
```
