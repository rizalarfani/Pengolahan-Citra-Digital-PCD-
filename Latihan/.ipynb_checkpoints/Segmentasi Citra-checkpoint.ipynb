# Segmentasi Warna

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import os

# Lokasi gambar anda
path = "../image/gambar.jpg" 
# Membaca gambar
image = cv2.imread(path)
# Bluring Gambar
image = cv2.medianBlur(image, (5))

# Batas Bawah warna hijau dan Batas Akhir Warna hijau
lower_green = np.array([0,100,0])
upper_green = np.array([80,255,100])

# Masking dan Segmentasi
mask = cv2.inRange(image,lower_green,upper_green)
segmentasi = cv2.bitwise_and(image,image,mask = mask)

#Menggunakan Metode cv2.imshow()
#untuk diplay gambar
cv2.imshow('Citra Asli',image)
cv2.imshow('Citra Segmentasi Warna',segmentasi)
cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```

# Segmentasi Sudut Bangun

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import os

# Lokasi gambar anda
path = "../image/gambar.jpg"
# Membaca gambar
image = cv2.imread(path)
image = cv2.medianBlur(image, (5))
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
th, im_th = cv2.threshold(gray, 200, 255, cv2.THRESH_BINARY_INV);
print(im_th)
circles = cv2.HoughCircles(im_th,cv2.HOUGH_GRADIENT,1,20,param1=50,param2=30,minRadius=0,maxRadius=0)
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        # draw the outer circle
        cv2.circle(image,(i[0],i[1]),i[2],(250,0,0),3)
    inv_cimg = cv2.bitwise_and(image,image)
    
plt.subplot(1, 3, 1), plt.imshow(image), plt.title("Gambar Asli")
plt.subplot(1, 3, 2), plt.imshow(inv_cimg), plt.title("Segmentasi Lingkaran")
```
