```python
import cv2
import os

# Menentukan directory vidio
lokasi_vidio = "../image/gelas.mp4"

# Menentukan variable unutk counter gambar
gambar_counter = 0

# Menentukan ukuran tinggi dan lebar dari gambar
tinggi = 200
lebar = 300
dim = (lebar,tinggi)

# Buat object vidio atau baca dari file input
vidio = cv2.VideoCapture(lokasi_vidio)

# Perikasa apakah vidio berhasil dibuka?
if vidio.isOpened() == False:
    print("Vidio gagal dibuka atau gagal dibaca")
    
while(vidio.isOpened()):
    # Tangkap frame per frame
    ret, frame = vidio.read()
    # Lakukan rizise gambar
    frame = cv2.resize(frame,dim)
    if ret == True:
        #Tampilkan vidio
        cv2.imshow("Frame",frame)
        
        k = cv2.waitKey(1)
        # ketika tombol Q ditekan maka akan berhenti vidio
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
        #Ketika tombol SPACE ditekan makan akan split gambar
        elif k%256 == 32:
            # tentukan folder untuk simpan gambar nya
            img_name = "../image/gelas/gelas_{}.png".format(gambar_counter)
            print(img_name)
            cv2.imwrite(img_name,frame)
            print('{} writen!'.format(img_name))
            #Gambar counter tambah 1
            gambar_counter += 1
vidio.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os

# Menentukan directory vidio
lokasi_vidio = "../image/gelas.mp4"

# Menentukan variable unutk counter gambar
gambar_counter = 0

# Menentukan ukuran tinggi dan lebar dari gambar
tinggi = 200
lebar = 300
dim = (lebar,tinggi)

list_frame = []

# Buat object vidio atau baca dari file input
vidio = cv2.VideoCapture(lokasi_vidio)

# Perikasa apakah vidio berhasil dibuka?
if vidio.isOpened() == False:
    print("Vidio gagal dibuka atau gagal dibaca")

print("____start load video____")
len_frames = int(vidio.get(cv2.CAP_PROP_FRAME_COUNT))
while(vidio.isOpened()):
    # Tangkap frame per frame
    ret, frame = vidio.read()
    # Lakukan rizise gambar
    frame = cv2.resize(frame,dim)
    if ret == True:
        #Tampilkan vidio
        if frame is None:
            continue
        list_frame.append(frame)
        if len_frames == len(list_frame):
            break
        
        # tentukan folder untuk simpan gambar nya
        img_name = "../image/gelas/gelas_{}.png".format(gambar_counter)
        print(img_name)
        cv2.imwrite(img_name,frame)
        print('{} writen!'.format(img_name))
        #Gambar counter tambah 1
        gambar_counter += 1
        cv2.imshow("Frame",frame)
        
        k = cv2.waitKey(1)
        # ketika tombol Q ditekan maka akan berhenti vidio
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
print("____complete load video____")         
vidio.release()
cv2.destroyAllWindows()
```
