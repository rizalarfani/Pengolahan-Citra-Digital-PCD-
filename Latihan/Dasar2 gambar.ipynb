# Image Rizise

```python
import cv2 # load pustaka openCv
# Lokasi gambar anda
path = "../image/gambar1.jpg" 
# Membaca gambar
image = cv2.imread(path,1)
# Scale
scale_percent = 10
# mengetahui ukuran image
height, width, channels = image.shape

height = int(height * scale_percent / 100)
width = int(width * scale_percent / 100 )
dim = (width,height)
resize = cv2.resize(image,dim,interpolation = cv2.INTER_AREA)

print("Citra Original: ", image.shape)
print("Citra Resize: ", resize.shape)
```

# Save file

```python
import cv2
import os
# Lokasi gambar
path = "../image/gambar1.jpg"
# Lokasi directori gambar
image_path = "../image/"
# Menggunakan Metode cv2.imread()
# untuk membaca gambar
image = cv2.imread(path,1)
# Ubah directory saat ini
# Ke directori yang ditentukan
os.chdir(image_path)
# Menggunakan Metode cv2.imwrite()
# untuk menyimpan sebuah file
cv2.imwrite("savedFile.jpg", image)
print("After saving image:")  
print(os.listdir(image_path))
print('Successfully saved')
```

# Membuat Garis
## dengan menggunakan cv2.line()

```python
import cv2 # load pustaka openCv
# Lokasi gambar anda
path = "../image/lena.jpg" 
# Membaca gambar
image = cv2.imread(path,1)
# Titik Awal (Koordinat X,Koordinat Y)
titik_awal = (0,0)
# Titik Akhir (Koordinat X,Koordinat Y)
titik_akhir = (image.shape[1],image.shape[0])
# Warna Garis/ niali BGR
color = (220,0,0)
# Ketebalan Garis dengan satuan PX
ketebalan = 9
# Membuat garis dengan motide cv2.line()
cv2.line(image,titik_awal,titik_akhir,(220,0,0),9)
cv2.imshow('Gambar 1',image)
cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```

# Membuat Lingkaran
## dengan menggunakan cv2.circle()

```python
import cv2 # load pustaka openCv
# Lokasi gambar anda
path = "../image/lena.jpg" 
# Membaca gambar
image = cv2.imread(path,1)
# Untuk memperoleh titik tengah
xTengah = int(image.shape[1] / 2)
yTengah = int(image.shape[0] / 2)
# Warna Garis/ niali BGR
color = (220,0,0)
# Ketebalan Garis dengan satuan PX
ketebalan = 5
# Membuat lingkaran dengan metode cv2.circle()
cv2.circle(image,(xTengah,yTengah),50,color,ketebalan)
# Menampilkan gambar
cv2.imshow("Lingkaran",image)
cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```

# Membuat Persegipanjang
## dengan menggunakan cv2.rectangle()

```python
import cv2 # load pustaka openCv
# Lokasi gambar anda
path = "../image/lena.jpg" 
# Membaca gambar
image = cv2.imread(path,1)
# Titik Awal sudut kiri atas persegi panjang
titik_awal = (5,5)
# Titik Akhir sudut kanan bawah persegi panjang
titik_akhir = (220, 220)
# Warna Garis/ niali BGR
color = (220,0,0)
# Ketebalan Garis dengan satuan PX
ketebalan = 2
# Membuat garis dengan motide cv2.rectangle()
cv2.rectangle(image,titik_awal,titik_akhir,color,ketebalan)
cv2.imshow('Persegi Panjang',image)
cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```

# Membuat Tulisan text
## dengan menggunakan cv2.putText()

```python
import cv2 # load pustaka openCv
# Lokasi gambar anda
path = "../image/lena.jpg" 
# Membaca gambar
image = cv2.imread(path,1)
# koordinat sudut kiri bawah teks pada gambar. 
# Koordinat ini nilai (nilai koordinat X, nilai koordinat Y). 
koordinat = (50,50)
# Warna Garis/ niali BGR
color = (220,0,0)
# Ketebalan Garis dengan satuan PX
ketebalan = 2
# Faktor skala font yang dikalikan dengan ukuran dasar font-spesifik
fontScale = 1
# jenis font text
font = cv2.FONT_HERSHEY_SIMPLEX
# Membuat tulisan text dengan motide cv2.putText()
# Using cv2.putText() method
image = cv2.putText(image, 'Text Tulisan', koordinat, font, 
                   fontScale, color, ketebalan, cv2.LINE_AA)
# menampikan gambar
cv2.imshow('Tulisan Text',image)
cv2.waitKey(0)  
# menutup semua jendela yang terbuka
cv2.destroyAllWindows() 
```
