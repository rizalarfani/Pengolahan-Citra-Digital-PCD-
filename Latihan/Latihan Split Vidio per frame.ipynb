```python
import cv2
import os

# Menentukan directory dari vidio
lokasi_vidio = "gelas.mp4"

# Menentukan variable untuk counter gambar
img_counter = 0

# Menentukan ukuran tinggi dan lebar dari gmabar
tinggi = 200
lebar = 300
dim = (lebar,tinggi)

# Buat object vidio atau baca dari file input
vidio = cv2.VideoCapture(lokasi_vidio)

# Periksa apakah vidio berhasil dibuka?
if vidio.isOpened() == False:
    print("Vidio gagal dibukan,Mohon cek lokasi vidio nya!!")


while(vidio.isOpened()):
    # Tangkap frame per frame
    ret, frame = vidio.read()
    # Lakukan resize gambar
    frame = cv2.resize(frame,dim)
    # Menampilkan vidio    
    cv2.imshow("Frame vidio", frame)
    
    k = cv2.waitKey(1)
    # Buat Kondisi ketika tombol Q di keyboard ditekan maka akan break atau stop video
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
    # Buat kondisi ketika tombol "SPACE" dikeyboard ditekan maka akan split gambar
    elif k%256 == 32:
        # Tentukan folder dan nama filenya
        img_name = "gelas/gelas_{}.png".format(img_counter)
        # Simpan ke gambar dengan cv2.imwrite
        cv2.imwrite(img_name,frame)
        print('{} writen!'.format(img_name))
        img_counter += 1

vidio.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os

# Menentukan directory dari vidio
lokasi_vidio = "gelas.mp4"

# Menentukan variable untuk counter gambar
img_counter = 0

list_frame = []

# Menentukan ukuran tinggi dan lebar dari gmabar
tinggi = 200
lebar = 300
dim = (lebar,tinggi)

# Buat object vidio atau baca dari file input
vidio = cv2.VideoCapture(lokasi_vidio)

# Periksa apakah vidio berhasil dibuka?
if vidio.isOpened() == False:
    print("Vidio gagal dibukan,Mohon cek lokasi vidio nya!!")

# Untuk mengambil jumlah frame
len_frame = int(vidio.get(cv2.CAP_PROP_FRAME_COUNT))
while(vidio.isOpened()):
    # Tangkap frame per frame
    ret, frame = vidio.read()
    # Lakukan resize gambar
    frame = cv2.resize(frame,dim)
    # Menampilkan vidio    
    cv2.imshow("Frame vidio", frame)
    
    if frame is None:
        continue
    list_frame.append(frame)
    if len_frame == len(list_frame):
        break
    
    # Tentukan folder dan nama filenya
    img_name = "gelas/gelas_{}.png".format(img_counter)
    # Simpan ke gambar dengan cv2.imwrite
    cv2.imwrite(img_name,frame)
    print('{} writen!'.format(img_name))
    img_counter += 1
    
vidio.release()
cv2.destroyAllWindows()
```
