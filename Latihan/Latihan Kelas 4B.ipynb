```python
# Menentukan lokasi atau folder dari citra yang akan digunakan
lokasi_gambar = "drive/MyDrive/Dataset Baru"

# Menentukan ukuran tinggi dan lebar pada citra
tinggi = 200
lebar = 200
dim = (lebar,tinggi)

# Buat variable penampung data dari citra
tampungan = []
x = []

# Mengumpulkan data citra yang akan dibuka
for path,sub_directory,nama_file in os.walk(lokasi_gambar):
  print(nama_file)
  for name in nama_file:
    print(name)
    # Gabungkan nama file dengan lokasi gambar
    img_path = (os.path.join(path,name))
    print(img_path)
    if img_path.endswith('png'):
      # Membaca gambar dari img_path
      gambar = cv2.imread(img_path)
      # Convert ke grayscale
      convert_to_gray = cv2.cvtColor(gambar,cv2.COLOR_BGR2GRAY)
      # Resize gambar
      resize_gambar = cv2.resize(convert_to_gray,dim)
      # untuk menampilkan gambar
      cv2_imshow(resize_gambar)
      # Blur
      blur = cv2.GaussianBlur(resize_gambar,(5,5),0)
      tampungan.append(blur.flatten())
x = np.array(tampungan)
print(x.shape)
```
