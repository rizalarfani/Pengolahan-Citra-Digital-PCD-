```python
# Nama: Mohamad Tegar Rahmawan
# NIM/Kelas: 20040186/4A

print("""
    *** KONSEP "SEDERHANA" GAME MOBIL ***
    ketik "help" untuk bantuan
    """)
perintah = ""
mulai = False
while True:
    perintah = input("-> ").upper()
    if perintah == "START":
        if mulai:
            print("Mobil telah berjalan !")
        else:
            mulai = True
            print("Mobil mulai berjalan... Siap untuk pergi !")
    elif perintah == "STOP":
        if not mulai:
            print("Mobil telah berhenti !")
        else:
            mulai = False
            print("Mobil berhenti !")
    elif perintah == "HELP":
        print("""
start -> untuk menjalankan mobil
stop -> untuk menghentikan mobil
exit -> untuk keluar dari program
        """)
    elif perintah == "EXIT":
        break
    else:
        print("Maaf, perintah tidak diketahui !")
```
