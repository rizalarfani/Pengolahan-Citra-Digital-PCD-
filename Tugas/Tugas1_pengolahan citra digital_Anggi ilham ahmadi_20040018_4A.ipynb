```python
print(40*"=")
print("   program sederhana by ilham ahamadi")
print("    silahkan pilih menu dibawah ini")
print(40*"=")
print("""1. Mencari luas persegi panjang
2. Mencari luas segitiga
3. Mencari luas lingkaran""")
print(40*"=")
pilih_menu =  input("masukan menu yang anda pilih : ")
print(40*"=")

if pilih_menu == "1":
    print("Program mencari luas persegi panjang")
    pplebar = float(input("masukan lebar persegi : "))
    pppanjang = float(input("masukan panjang persegi : "))
    ppluas = pplebar * pppanjang
    print(40*"=")
    print("luas persegi panjang adalah ", ppluas)
    
elif pilih_menu == "2" :
    print("Program mencari luas segitiga")
    sgtinggi = float(input("masukan tinggi segitiga : "))
    sgpanjang = float(input("masukan panjang segitiga : "))
    sgluas = 0.5 * sgtinggi * sgpanjang
    print(40*"=")
    print("luas segitiga adalah ", sgluas)
    
elif pilih_menu == "3":
    print("Program mencari luas lingkaran")
    lgjari = float(input("masukan panjang jari - jari : "))
    lgluas = 3.14 * lgjari * lgjari
    print(40*"=")
    print("luas lingkaran adalah ", lgluas)
    
else : print("pilihan salah")
print(40*"=")
```
